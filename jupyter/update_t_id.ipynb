```python
import firebase_admin
from firebase_admin import credentials, firestore

# Initialize the Firebase app with service account credentials
cred = credentials.Certificate("C:\\Users\\AKHNA\\Downloads\\smart-tour-guide-3d30d-firebase-adminsdk-h17r2-bda949eabd.json")  # Replace with the path to your JSON key file
firebase_admin.initialize_app(cred)
db = firestore.client()

def update_reviews_with_tourist_ids():
    try:
        # Reference to the 'tourist' and 'reviews' collections
        tourist_ref = db.collection('tourist')
        reviews_ref = db.collection('reviews')
        
        # Cache tourist IDs for faster lookup
        tourist_ids = {doc.id: doc.to_dict()['t_id'] for doc in tourist_ref.stream()}
        
        # Update each review document's 't_id' to match the tourist document ID
        for review in reviews_ref.stream():
            review_data = review.to_dict()
            current_t_id = review_data.get('t_id')
            
            # Find the correct document ID from the tourist cache
            tourist_id = next((doc_id for doc_id, t_id in tourist_ids.items() if t_id == current_t_id), None)
            
            if tourist_id:
                # Update the 't_id' field in the review to the new tourist document ID
                reviews_ref.document(review.id).update({'t_id': tourist_id})
                print(f"Updated review ID: {review.id} with new t_id: {tourist_id}")
            else:
                print(f"No matching tourist document ID found for t_id: {current_t_id} in review ID: {review.id}")
        
        print("All review t_id fields updated successfully.")
    
    except Exception as e:
        print(f"An error occurred: {e}")

# Run the function
update_reviews_with_tourist_ids()
```
